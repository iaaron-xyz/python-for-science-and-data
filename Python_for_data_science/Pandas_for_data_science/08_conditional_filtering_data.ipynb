# Conditional filtering

Windows

```python
# Import the libraries
import os
import pandas as pd

# Read the data from a CSV file
# Get the absolute path to the current notebook
os_path = os.getcwd()
# Add the extra path to the dataset file
dataset_path = os_path+'\datasets\Online_Retail.csv'
retail_data = pd.read_csv(dataset_path, encoding='ISO-8859-1')

# Convert the column "InvoiceDate" to a datetime type
retail_data['InvoiceDate'] = pd.to_datetime(retail_data['InvoiceDate'])

# Delete the rows in important columns with empty values
retail_data.dropna(subset=['CustomerID', 'InvoiceDate'], inplace=True)

# Create a new column 'TotalPrice'
retail_data['TotalPrice'] = retail_data['Quantity'] * retail_data['UnitPrice']

retail_data
```

## Filter for a specific element

Filter only the sales made to Malta

```python
malta_sales = retail_data[retail_data['Country'] == 'Malta']
malta_sales
```

## Filter by quantity

```python
quantity_20 = retail_data[retail_data['Quantity'] == 20]
quantity_20
```

## Filter using Multiple Conditions

You can combine multiple boolean conditions using the pandas' logic operators:

- `&` para *AND*.

- `|` para *OR*.

- `~` para *NOT*.

Remember always enclose in parenthesis `()` every boolean condition.

Get only the rows with France as a country and with exactly 20 sales.

```python
retail_fra_greater_20 = retail_data[(retail_data['Quantity'] == 20) & (retail_data['Country'] == 'France')]
retail_fra_greater_20
```

## Filter by Year

Get the rows made in the year 2010

```python
sales_2010 = retail_data[retail_data['InvoiceDate'].dt.year == 2010]
sales_2010
```

Get the sales made in the year 2010 only in Germany

```python
sales_germany_2010 = retail_data[
    (retail_data["InvoiceDate"].dt.year == 2010) & (retail_data["Country"] == "Germany")
]
sales_germany_2010
```

## Filter by year and month

Get the sales made in the UK in the year 2011 in the month of December

```python
sales_uk_2011 = retail_data[
    (retail_data["InvoiceDate"].dt.year == 2011) & (retail_data["InvoiceDate"].dt.month == 12) & (retail_data["Country"] == "United Kingdom")
]
sales_uk_2011
```

## Filter by using the `isin()` method

This method is iseful to find elements within a column or Series that match with specific values.

```python
filter_isin = retail_data[retail_data['Country'].isin(['Spain', 'Monaco'])]
filter_isin
```

## Filter empty or null values

In this example, there are not empty (`None`, `null`, etc) values.

```python
filter_nulls = retail_data[retail_data['Description'].isnull()]
filter_nulls
```

## Filter complex conditions using the `query()` method

Pandas provides the `query()` method to filter using conditions wit SQL similar syntaxis.

```python
# Filter rows where the TotalPrice is bigger that 400.00 and Quantity is less than 30
filter_query_totalprice = retail_data.query('TotalPrice > 400 and Quantity < 30')
filter_query_totalprice
```

## Filter by text in strings using `str.contains()`

If you need to find a substring in a chain of text, then the method for you is `str.contains()` to search patterns in columns of type `string`.

In this example, search for all rows that contain the sub-string `SPACE` in the **Decription**.

```python
filter_strcontains = retail_data[retail_data['Description'].str.contains('SPACE')]
filter_strcontains
```

## Filter using indexes

To filter using the indexes of a row you can use `loc[]` and `iloc[]`. Check the next example:

```python
# Get only the indexes 2 and 43
filter_index_0and43 = retail_data.loc[[2,43]]
filter_index_0and43
```

```python
# Get the indexes 35 to 43
filter_index_35to43 = retail_data.iloc[35:44]
filter_index_35to43
```

## Filter only unique rows using `drop_duplicates()`

The `drop_duplicates()` is ised to filter the duplicated rows.

```python
filter_duplicates = retail_data.drop_duplicates()
filter_duplicates
```

Remove duplicated rows only based on specific columns, apply it to countries to get only a different country on each row.

```python
filter_duplicates_row = retail_data.drop_duplicates(subset='Country')
filter_duplicates_row
```

## Filter by a range of values using `between()`

The method `between()` allows to filter within a given range. In the next example, find only the uitary prices that are between 20 and 30.

```python
filter_between = retail_data[retail_data['UnitPrice'].between(20,30)]
filter_between
```
