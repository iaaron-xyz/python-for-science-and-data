# Pivot tables and Reshaping in Pandas

## Pivot tables in Pandas

### Creating a pivot table with `pivot_table()`

In this example, we want to know what is the total quantity of every retail element given its description per Country.

```python
pivot_1 = pd.pivot_table(retail_data, values='Quantity', index='Country', columns='Description', aggfunc='sum')
pivot_1
```

## Reshaping in Pandas

### `stack()` and `unstack()`

```python
# Original DataFrame
df
```

```python
# Stack a Dataframe
df_stacked = df.stack()
df_stacked.reset_index()
```

```python
# Unstack a stacked dataframe
df_unstacked = df_stacked.unstack()
df_unstacked
```
