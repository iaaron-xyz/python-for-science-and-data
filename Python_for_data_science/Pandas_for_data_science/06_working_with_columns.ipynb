# Creating and Manipulating columns

## Import libraries and datasets

```python
import os
import pandas as pd

# Get the absolute path to the current notebook
os_path = os.getcwd()
# Add the extra path to the dataset file
dataset_path = os_path + '\datasets\Online_Retail.csv'
retail_data = pd.read_csv(dataset_path, encoding='ISO-8859-1')
retail_data
```

## Create a new column from the existent ones

### Multiply 2 columns and get a 3rd as a result

```python
retail_data['TotalPrice'] = retail_data['Quantity'] * retail_data['UnitPrice']
retail_data
```

### Apply comparison operators to a columns

```python
retail_data['HighUPrice'] = retail_data['UnitPrice'] > 3
retail_data
```

### Delete a column

Delete the recently created column

```python
retail_data.drop('HighUPrice', axis=1, inplace=True)
retail_data
```

## Get the info data type of every column with the method `.info()`

```python
retail_data.info()
```

## Convert the datatype of a column

Convert the Column `InvoiceDate` from `object` to `datetime` type.

```python
retail_data['InvoiceDate'] = pd.to_datetime(retail_data['InvoiceDate'])
retail_data['Description'] = retail_data['Description'].astype('string')
retail_data.info()
```

## Use `.apply(func)`

### Create new columns using lambda functions

```python
retail_data['Discount10'] = retail_data['UnitPrice'].apply(lambda x: x*0.9)
retail_data
```

### create categories using `apply` and a custom function

```python
# Custom function to categorize
def categorize_price(price):
    if price > 5:
        return 'High'
    elif price > 3:
        return 'Medium'
    else:
        return 'Low'
    
retail_data['UPriceCategory'] = retail_data['UnitPrice'].apply(categorize_price)
retail_data.head(20)
```

## Get the list of countries

```python
countries = retail_data['Country'].value_counts()
countries
```
