# Basic plot: The first plot in matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Line plots

### My first plot

```python
# y axis
months = np.array(["Jan", "Feb", "Mar", "Apr", "May"])
# x axis
sales = np.random.randint(50, 200, 5)

# Set the graphic size
plt.figure(figsize=(8, 6))

# Define the graphic style
plt.plot(months, sales, marker="o", color="blue")

# Add title and labels
plt.title("Monthly sales")
plt.xlabel("Months")
plt.ylabel("Sales (in thousands)")

# Show the plot
plt.show()
```

### Math plot

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)


# Crear la figura y los ejes
plt.figure(figsize=(8, 6))

# Graficar la lnea
plt.plot(x, y, label='Sin(x)', color='g', linestyle='-', linewidth=2)

# Show plot
plt.show()
```

## Dispersion plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# x axis
hours = np.linspace(1,8,8)
# y axis
exam = [int(np.random.randint(i-2, i+2, 1)[0]) for i in hours]

# Size of the plot
plt.figure(figsize=(10,8))
# Type of plot
plt.scatter(
  x=hours,
  y=exam,
  marker="o"
)

# Title and labels
plt.title("Relation between study hours and grades")
plt.xlabel("Study hours")
plt.ylabel("Qualifications obtained")

# Show the plot
plt.show()
```
