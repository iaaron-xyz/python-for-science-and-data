# Pie charts and Bar charts

````python
import matplotlib.pyplot as plt
import numpy as np
````

## Barchart

### Vertical barchart

````python
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales = [120, 150,90]

# Making the barchart
plt.bar(categories, sales, color='skyblue', label='Mothly sales', width=0.99)

# title and labels
plt.title('Sales by month')
plt.xlabel('Products')
plt.ylabel('Sales')

# Add legent
plt.legend()

# Add annotation
plt.annotate(
    'Lower income',
    xy=('Product C', 93),
    xytext=(1.8, 110),
    arrowprops=dict(facecolor='teal')
)

# render the plot
plt.show()
````

### Horizontal barchart

````python
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales = [120, 150,90]

# Making the barchart
plt.barh(categories, sales, color='skyblue', label='Mothly sales')

# title and labels
plt.title('Sales by month')
plt.xlabel('Sales')
plt.ylabel('Products')

# Add legent
plt.legend()

# Add annotation
plt.annotate(
    'Lower income',
    xy=(93, 'Product C'),
    xytext=(110, 1.8),
    arrowprops=dict(facecolor='teal')
)

# render the plot
plt.show()
````

## Stacked barchart

````python
plt.figure(figsize=(8,6))

categories = ['Product A', 'Product B', 'Product C']
sales1 = [120, 150,90]
sales2 = [100, 120, 200]

# Making the barchart
plt.bar(categories, sales1, color='skyblue', label='Electronics')
plt.bar(categories, sales2, bottom=sales1, label='Software')

# title and labels
plt.title('Stacked bars plot')
plt.xlabel('Sales')
plt.ylabel('Products')

# Add legent
plt.legend()

# render the plot
plt.show()
````

## Piechart

### Basic piechart

````python
plt.figure(figsize=(8,6))

total = 360
categories = ['Product A', 'Product B', 'Product C']
sales = [(120/total)*100, (150/total)*100, (90/total)*100]

# Making the piechart
plt.pie(sales, labels=categories, autopct='%1.1f%%', startangle=140, colors=['dodgerblue', 'hotpink', 'lawngreen'])

# title and labels
plt.title('Sales by month')

# plot as a circle
plt.axis('equal')

# render the plot
plt.show()
````

### Add custom elements

In this section, it will be shown how a piechart can be customized with paramaters like, `explode`, `autopct`, `shadow`, and add a gradient of colors using `plt.cm.viridis()`.

To highlight a section or segment in a pie chart use `explode`, this paramater allows to select what segment and ho much to separate it from the original position:

```python
# Exploding
explode = (0.1, 0, 0)
plt.pie(sales, labels=categories, explode=explode)
```

The position is the same as in the labels.

````python

# Exploding: highlight the category
explode = (0.1, 0, 0)  # Highlight the first segment

# Add a gradient of colors
colors = plt.cm.cool(np.linspace(0, 0.7, len(sales)))

# Create piecharts showing percentages
plt.pie(sales, labels=categories, colors=colors, autopct='%1.1f%%', explode=explode, shadow=True, startangle=100)

# Ensure the chart is circular
plt.axis('equal')

# Add title
plt.title('Diagrama de Pastel con Porcentajes')

# Render the chart
plt.show()
````

## A barchart in polar coordinates

````python
# Generate random data
np.random.seed(19680801)

# Define the parameters for the polar bars with an angle theta
N = 5
theta = np.linspace(0.0, 2 * np.pi, N, endpoint=False)
# height/radius of the bars
radii = 10 * np.random.rand(N)
# width of the bars
width = np.pi / 4 * np.random.rand(N)
# color variation depending on the radio/height
colors = plt.cm.viridis(radii / 10.)

# Plot the barchart in polar coordinates
ax = plt.subplot(projection='polar')
ax.bar(theta, radii, width=width, bottom=0.0, color=colors, alpha=0.7)

# render the plot
plt.show()
````
