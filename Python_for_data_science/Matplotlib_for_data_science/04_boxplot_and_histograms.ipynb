# Histograms and Boxplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Set a seed
np.random.seed(1572019)

# Create the data set
data = np.random.normal(0, 10, 200)
data
```

## Histogram

### Basic histogram

```python
# number of intervals
intervals = 10


plt.hist(data, color='skyblue', edgecolor='skyblue', alpha=0.5)

plt.title('Histogram')
plt.xlabel('X axis')
plt.ylabel('Y Axis')

plt.show()
```

### Custom histogram

```python
# Number of intervals
intervals = 10

# Create histogram data
counts, bins = np.histogram(data, bins=intervals)

# Create a color gradient using the 'cool' colormap
colors = plt.cm.cool(np.linspace(0, 1, intervals))

# Create the histogram with color gradient
for count, x in zip(counts, bins[:-1]):
    plt.bar(x, count, width=bins[1] - bins[0], color=colors[counts.tolist().index(count)], edgecolor='grey', alpha=0.6)

plt.title('Histogram with Color Gradient')
plt.xlabel('X axis')
plt.ylabel('Y Axis')
```

### 2D Histograms

2D histograms are like seeing a 2D histogram from the top. In this example, the darker the color, the bigger the bars.

```python
mean = [10, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T

# Histogram 2D
plt.hist2d(x, y, bins=30, cmap='Blues')
plt.colorbar()
plt.show()
```

It is also a hexagons version available, the concept is the same

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar()
plt.show()
```

## Box Plots

### Basic boxplots

```python
# Generate the data
np.random.seed(33)

# data
ages = [
    np.random.normal(20, 5, 100),
    np.random.normal(30, 5, 100),
    np.random.normal(40, 5, 100),
]

ages
```

```python
# Plot

plt.boxplot(ages, patch_artist=False, notch=False, vert=True, tick_labels=['Young', 'Young-Adult', 'Adult'])

plt.title('Age distribution')
plt.xlabel('Group')
plt.ylabel('Age')

plt.show()
```
