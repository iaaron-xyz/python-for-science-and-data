# Plot customization

````python
import numpy as np
import matplotlib.pyplot as plt
````

## Basic line plot

````python
# x axis
x = np.linspace(0, 6*np.pi)
# y axis
y = np.cos(x)

# Size of the plot
plt.figure(figsize=(10, 8))
# Type of plot
plt.plot(x, y)

# Title and labels
plt.title("Cosine")
plt.xlabel("x")
plt.ylabel("y = Cos(x)")

# Show the plot
plt.show()
````

### Colors and Styles of line

You can define the width and the type of the line in line plots.

Available styles
- `-` : línea sólida
- `--` : línea discontinua
- `:` : línea punteada

```python
plt.plot(x, y, color='green', linestyle='--', linewidth=2)
```

You can define colors in matplotlib:

- Named colors (e.g., `red`, `blue`, `green`)
- Hexadecimal color codes (e.g., `#FF5733`)
- RGB/RGBA tuples (e.g., `(1, 0, 0)`for red)
- Colormaps (e.g., `cmap='viridis'`)

The full list of colors can accesed in the [Oficial site: List of named colors](https://matplotlib.org/stable/gallery/color/named_colors.html#sphx-glr-gallery-color-named-colors-py)

#### Plot

````python
# x axis
x = np.linspace(0, 6 * np.pi)
# y axis
y = np.cos(x)

# Size of the plot
plt.figure(figsize=(10, 8))

# Type of plot
plt.plot(x, y, color='fuchsia', linestyle=':', linewidth=2)

# Title and labels
plt.title("Cosine", fontsize=21, color="teal")
plt.xlabel("x")
plt.ylabel("y = Cos(x)")

# axes limits
plt.xlim(-1, 20)
plt.ylim(-2, 2)

# axis ticks
plt.xticks(
    [0, np.pi, 2 * np.pi, 3 * np.pi, 4 * np.pi, 5 * np.pi, 6 * np.pi],
    ["0", "pi", "2*pi", "3*pi", "4*pi", "5*pi", "6*pi"],
    rotation=45,
)

# Show the plot
plt.show()
````

### Anotations

Add annotations in specific parts of the text.

```python
matplotlib.pyplot.annotate(
    text,         # The text of the annotation
    xy,           # The point (x, y) to annotate
    xytext=None,  # Optional: Position (x, y) for the text
    xycoords='data',  # Coordinate system for xy
    textcoords=None,   # Coordinate system for xytext
    arrowprops=None,   # Optional: Properties for the arrow
    **kwargs         # Additional keyword arguments for styling
)
```

#### Plot

````python
# x axis
x = np.linspace(0, 6 * np.pi)
# y axis
y1 = np.cos(x)
y2 = np.cos(x + np.pi/2)

# Size of the plot
plt.figure(figsize=(10, 8))

# Type of plot
plt.plot(x, y1, color='seagreen', linestyle='--', linewidth=2, label='Cos(x)')
plt.plot(x, y2, color='orchid', linestyle='--', linewidth=2, label='Cos(x + pi/2)')

# Title and labels
plt.title("Cosine", fontsize=21, color="teal")
plt.xlabel("x")
plt.ylabel("y = Cos(x)")
plt.legend(loc='upper right')

# axes limits
plt.xlim(-1, 20)
plt.ylim(-2, 2)

# axis ticks
plt.xticks(
    [0, np.pi, 2 * np.pi, 3 * np.pi, 4 * np.pi, 5 * np.pi, 6 * np.pi],
    ["0", "pi", "2*pi", "3*pi", "4*pi", "5*pi", "6*pi"],
    rotation=45,
)

# Add annotation
plt.annotate(
    'Offset by one pi',
    xy=((7/4)*np.pi, 1),
    xytext=(2*np.pi, 1.3),
    arrowprops=dict(facecolor='teal')
)

# Show the plot
plt.show()
````

### Subplots

You can divide the plot total area into multiple smaller plots

```python
fig, ax = plt.subplots(2, 2)  # Grid of 2x2 plots
ax[0, 0].plot(x, y)
ax[1, 1].plot(x, z)
```

#### Plot

````python
# x axis
x = np.linspace(0, 6 * np.pi)
# y axis
y1 = np.cos(x)
y2 = np.sin(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Size of the plot
fig, ax = plt.subplots(2,2, figsize=(10,8))

# Subplot 1
ax[0,0].plot(x, y1, color='seagreen', linestyle='--', linewidth=2, label='Cos(x)')
ax[0,0].set_title('Cos(x)')
ax[0,0].set_title("Cosine", fontsize=21, color="teal")
ax[0,0].set_xlabel("x")
ax[0,0].set_ylabel("y = Cos(x)")
ax[0,0].legend(loc='upper right')
ax[0,0].set_xlim(-1, 20)
ax[0,0].set_ylim(-2, 2)
# axis ticks
ax[0,0].set_xticks(
    [0, np.pi, 2 * np.pi, 3 * np.pi, 4 * np.pi, 5 * np.pi, 6 * np.pi],
    ["0", "pi", "2*pi", "3*pi", "4*pi", "5*pi", "6*pi"],
    rotation=45,
)

ax[0,1].plot(x, y2, color='orchid', linestyle='--', linewidth=2, label='Sin(x)')
ax[0,1].set_title('Sin(x)')

ax[1,0].plot(x, y3, color='green', linestyle='-', label='Tan(x)')
ax[1,0].set_title('Tan(x)')

ax[1,1].plot(x, y4, color='dodgerblue', linestyle='--', linewidth=2, label='Exp(x)')
ax[1,1].set_title('exp(x)')

# Show the plot
plt.tight_layout()
plt.show()
````
