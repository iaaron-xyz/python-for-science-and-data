# Subplots and gridspec

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
```

```python
# Data
x = np.linspace(0, 10*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

## Subplots with `plt.subplot()` and `plt.subplots()`

### Basic subplots

```python
# A figure with 2 subplots (1 row, 2 columns)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot1: Sine
ax1.plot(x, y1, 'r')
ax1.set_title("Sin(x)")

# Plot 2: Cosine
ax2.plot(x, y2, 'b')
ax2.set_title("Cos(x)")

plt.show()
```

### Basic custom

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot 1: Sine
ax1.plot(x, y1, 'g')
ax1.set_title("Sin(x)")
ax1.set_xlabel("Time")
ax1.set_ylabel("Amplitude")


# Plot 2: Cosine
ax2.plot(x, y2, 'pink')
ax2.set_title("Cos(x)")
ax2.set_xlabel("Time")
ax2.set_ylabel("Amplitudw")

plt.show()
```

### Adjust space

```python
# increase the width size
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# plots
ax1.plot(x, y1, 'r')
ax2.plot(x, y2, 'b')

# Adjust the horizontal space
plt.subplots_adjust(wspace=0.9)

plt.show()
```

## Subplots with custom size with gridspace

### Add 3 subplots

```python
# gridspace custom matrix of 2 by 2
# [ O O ]
# [ 0 0 ]
gs = gridspec.GridSpec(3, 2, height_ratios=[1,1,1], width_ratios=[1,1])
fig = plt.figure(figsize=(12,8))

# Plots

# Plot top
ax1 = fig.add_subplot(gs[2,:])
ax1.plot(x, y1 + y2, color='limegreen')
ax1.set_title('Sin(x) + Cos(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('f(x)')
ax1.set_ylim(-2,2)

# Plot bottom left
ax2 = fig.add_subplot(gs[0,:])
ax2.plot(x, y1, color='skyblue')
ax2.set_title('Sin(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('Sin(x)')
ax2.set_ylim(-2,2)

# Plot bottom right
ax3 = fig.add_subplot(gs[1,:])
ax3.plot(x, y2, color='gold')
ax3.set_title('Cos(x)')
ax3.set_xlabel('x')
ax3.set_ylabel('Cos(x)')
ax3.set_ylim(-2,2)

# Render plot
plt.tight_layout()
plt.show()
```

### Share axis

Sometimes is useful create many subplots that share the same axis. Specially when you eant to compare plots and their scales.

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), sharey=True)


# Plot 1: Sin(x)
ax1.plot(x, y1, 'r')
ax1.set_title("Sin(x)")


# Plot 2: Cos(x)
ax2.plot(x, y2, 'b')
ax2.set_title("Cos(x)")


# Both subplots share the same y-axis
plt.show()
```

## Subplots with `plt.subplot2grid()`

```python
fig = plt.figure(figsize=(10, 7))

# Placing the plots in a 3x3 grid
ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)  # 2 columns
ax2 = plt.subplot2grid((3, 3), (0, 2))  # 3rd column only
ax3 = plt.subplot2grid((3, 3), (1, 0), colspan=3)  # full 2nd row
ax4 = plt.subplot2grid((3, 3), (2, 0), colspan=3)  # full 3rd row

# Plots
ax1.plot(x, y1, 'r')
ax1.set_title("Seno")

ax2.plot(x, y2, 'g')
ax2.set_title("Coseno")

ax3.plot(x, y1 + y2, 'b')
ax3.set_title("Suma de Seno y Coseno")

ax4.plot(x, y1 - y2, 'purple')
ax4.set_title("Diferencia entre Seno y Coseno")

plt.tight_layout()
plt.show()
```

## `plt_tight_layout()` and `plt.constrained_layout()`

Those are usefult to adjust the margins automatically, making sure labels and titles does not overlap.

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 4))


# Plots
ax1.plot(x, y1, 'r')
ax2.plot(x, y2, 'g')
ax3.plot(x, y1 + y2, 'b')


# Adjust margins automatically
plt.tight_layout()  
# plt.constrained_layout()
plt.show()
```
